```python
source = pd.read_csv(r"D:\00_BACKUP_BY_DRIVE\erdinctasel\erdinc.csv")


source['start'] = pd.to_datetime(source['start'],yearfirst=True,format="%Y-%m-%d")
source['end'] = pd.to_datetime(source['end'],yearfirst=True,format="%Y-%m-%d")
source['diff'] = source['end'] - source['start']
 
# Declaring a figure "gnt"
fig, ax = plt.subplots(figsize=(15,6))

ax.autoscale()
ax.set(title="Experience Timeline")
# Need to fix hidden tick labels
# https://stackoverflow.com/questions/43673659/matplotlib-not-showing-first-label-on-x-axis-for-the-bar-plot
source= source.sort_values(by='start').reset_index(drop=True).reset_index()
ax.set_xlim(xmin=1995, xmax=2030)
y_tick_labels = source.position.values
y_pos = np.arange(len(y_tick_labels))

ax.set_yticks(y_pos)
ax.set_yticklabels(y_tick_labels)

for index, row in source.iterrows():
    start_year = int(row.start.strftime("%Y"))+int(row.start.strftime("%m"))/12 + int(row.start.strftime("%d"))/30
    duration = row['diff'].days/365
    ax.broken_barh([(start_year, duration)], 
                    (index-0.5,0.8), 
                    facecolors =('tan'),
                    label=row.experience
                  )
    ax.text(start_year+0.5, index-0.2, row.experience)
```

```python
source = pd.read_csv(r"D:\00_BACKUP_BY_DRIVE\erdinctasel\erdinc.csv")


source['start'] = pd.to_datetime(source['start'],yearfirst=True,format="%Y-%m-%d")
source['end'] = pd.to_datetime(source['end'],yearfirst=True,format="%Y-%m-%d")
source['diff'] = source['end'] - source['start']
source= source.sort_values(by='start').reset_index(drop=True)
dates_start=source["start"]
xp= source["experience"]
position= source["position"]

# Choose some nice levels
levels = np.tile([-5, 5, -3, 3, -1, 1],
                 int(np.ceil(len(dates_start)/6)))[:len(dates_start)]

# Create figure and plot a stem plot with the date
fig, ax = plt.subplots(figsize=(20, 5), layout="constrained")
ax.set(title="Experience Starting Timeline")

ax.vlines(dates_start, 0, levels, color="tab:green")  # The vertical stems.
ax.plot(dates_start, np.zeros_like(dates_start), "-o",
        color="r", markerfacecolor="w")  # Baseline and markers on it.


# annotate lines
for d, l, x in zip(dates_start, levels, xp):
    ax.annotate(x, xy=(d, l),
                xytext=(-3, np.sign(l)*3), textcoords="offset points",
                horizontalalignment="center",
                verticalalignment="bottom" if l > 0 else "top")

for d, l, p in zip(dates_start, levels, position):
    ax.annotate(p, xy=(d, l),
                xytext=(-3, np.sign(l)*13), textcoords="offset points",
                horizontalalignment="center",
                verticalalignment="top" if l < 0 else "bottom")
    
# format x-axis with 12-month intervals
ax.xaxis.set_major_locator(mdates.YearLocator(1,8,1))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
ax.xaxis.set_minor_locator(mdates.MonthLocator(interval=6))
ax.xaxis.set_minor_formatter(mdates.DateFormatter("%b"))

plt.setp(ax.get_xticklabels(which="both"), rotation=90, ha="center")

# remove y-axis and spines
ax.yaxis.set_visible(False)
#ax.spines[["left", "top", "right"]].set_visible(False)

ax.margins(y=0.1)
plt.show()
```
